# Basic Statistics Inspection

```python
import os
import pandas as pd
import csv
```

- data extraction

```python
CSVFiles_dir = "CSVFiles"

CSV_arrays = []

data_dic = {}

for filename in os.listdir(CSVFiles_dir):
    if filename.endswith(".csv"):
        data_dic[f"{filename}"] = {}
        file_path = os.path.join(CSVFiles_dir, filename)
        with open(file_path,  newline='') as f:
            reader = csv.reader(f)
            data = [row for row in reader]
            header = data[0]
            data = data[1:]  # Skip header row
            for i in range(len(data)):
                data_dic[f"{filename}"][f"{data[i][0]}"]= data[i]
                #for x in range(len(data[0])):
                    #data_dic[f"{filename}"][f"{data[i][0]}"][f"{header[x]}"] = {data[i][x]}
```

```python
df = pd.DataFrame(data_dic)
```
